```
import numpy as np
import matplotlib.pyplot as plt
```

# Vector Plotting

```
plt.quiver(0,0,3,4)
plt.show()
```

```
plt.quiver(0,0,3,4, scale_units='xy', angles='xy', scale=1)
plt.show()
```

```
plt.quiver(0,0,3,4, scale_units='xy', angles='xy', scale=1)
plt.xlim(-5, 5)
plt.ylim(-5,5)
plt.show()
```

```
plt.quiver(0,0,3,4, scale_units='xy', angles='xy', scale=1, color='red')
plt.quiver(0,0,-3,4, scale_units='xy', angles='xy', scale=1, color='green')
plt.xlim(-5, 5)
plt.ylim(-5,5)
plt.show()
```

```
def plot_vectors(vecs):
  for vec in vecs:
    plt.quiver(vec[0], vec[1], vec[2], vec[3], scale_units='xy', angles='xy', scale=1, color='red')
  plt.xlim(-10, 10)
  plt.ylim(-10,10)
  plt.show()
```

```
plot_vectors([(0,0,3,4), (0,0,-3,4), (0,0,-2,-5), (0,0,1,-5)])
```

# Vector addition/ subtraction

```
vecs = [np.array([0,0,3,4]), np.array([0,0,-3,4]), np.array([0,0,-2,-5]), np.array([0,0,1,-5])]
```

```
plot_vectors([(0,0,3,4), (0,0,1,-5)])
```

```
vecs[0] + vecs[3]
```

```
plot_vectors([vecs[0], vecs[3], vecs[0]+vecs[3]])
```

```
plot_vectors([vecs[0], vecs[3], vecs[0]-vecs[3]])
```
